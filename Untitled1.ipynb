```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("insurance_pre.csv")
data
```

```python
data =pd.get_dummies(data,drop_first=True)
data
```

```python
data['sex_male']=data['sex_male'].replace({True:1,False:0})
data
```

```python
data['smoker_yes']=data['smoker_yes'].replace({True:1,False:0})
data
```

```python
independent= data[["age","bmi","children","sex_male","smoker_yes"]]
dependent =data[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_
print("weight of the model = {}".format(weight))
```

```python
bais=regressor.intercept_
print("bais of the model={}".format(bais))
```

```python
y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
